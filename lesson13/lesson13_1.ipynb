```python
import matplotlib.pyplot as plt

# 1.製作figure
fig = plt.figure()

# 2.在1.的figure上製作一個axes
ax_1 = fig.add_subplot(1, 1, 1)

# 3.將圖的數據儲存在axes中
X = [0, 1, 2]
Y = [3, 4, 5]
ax_1.plot(X, Y)

# 4.顯示圖
plt.show()
```

```python
import matplotlib.pyplot as plt

fig = plt.figure()

axes = fig.add_subplot(1,1,1)
x = [0, 1, 2]
y = [3, 4, 5]

axes.plot(x, y)

plt.show()
```

```python
fig = plt.figure()
axes1 = fig.add_subplot(1, 3, 1)
axes2 = fig.add_subplot(1, 3, 2)
axes3 = fig.add_subplot(1, 3, 3)

axes1.set_title('label1')
axes2.set_title('label2')
axes3.set_title('label3')

plt.show()
```

```python
import matplotlib.pyplot as plt
student1 = {'chinese':78, 'english': 80, 'history':75, 'discover': 85}
student2 = {'chinese':69, 'english': 92, 'history':69, 'discover': 80}

subjects = list(student1.keys())
scores1 = list(student1.values())
scores2 = list(student2.values())

fig = plt.figure()
axes = fig.add_subplot(1,1,1)
axes.plot(subjects, scores1, label='stu1')
axes.plot(subjects, scores2, label='stu2')
axes.legend()
plt.show()
```
